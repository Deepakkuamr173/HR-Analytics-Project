```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

file = "Attendance-Sheet-2022-2023.xlsx"

sheets = ["Apr 2022", "May 2022", "June 2022"]

all_data = []

for sheet in sheets:

    df = pd.read_excel(file, sheet_name=sheet)

    # Rename columns
    df.rename(columns={
        "AtliQ": "Employee Code",
        "Unnamed: 1": "Name"
    }, inplace=True)

    # Remove unwanted columns
    remove_cols = [
        'Total Present Days',
        'Present',
        'Work from home ',
        'Paid Leave ',
        'Sick Leave ',
        'Birthday Leave ',
        'Floting festival leave ',
        'Bereavement Leave',
        'Leave without pay',
        'Weekly Off',
        'Holidays Off',
        'Menstrual Leave'
    ]

    df = df.drop(columns=remove_cols, errors='ignore')

    #   # Remove duplicate header rows
    df = df[df["Employee Code"] != "Employee Code"]

    # Convert wide to long
    df = df.melt(
        id_vars=["Employee Code", "Name"],
        var_name="Date",
        value_name="Status"
    )

    # Add month
    df["Month"] = sheet

    all_data.append(df)

# Combine all sheets
final_df = pd.concat(all_data, ignore_index=True)

final_df
```
